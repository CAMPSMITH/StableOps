## What Happened

The profitable arbitrage opportunity in the Terra collapse was not from the volatility of the stablecoin price off of its peg and waiting for it to return to its peg value, UST/USD pair, instead it was from the inherent structure of the Terra ecosystem and the unique relationship between UST and LUNA which created a situation where instantaeous profitable arbitrage could be executed the instant UST fell below its peg value, UST / LUNA pair.

In the Terra ecosystem, the variable crypto currency **LUNA** is used to ensure that the **UST** maintains its pegged value.  1 **UST** can be exchanged with \\$1.00 worth of **LUNA**.  When the **UST** fell below its pegged value of \\$1.00, a massive amount or arbitrage occured which eventually led to the collapse ofboth **UST** and **LUNA**.  It can be seen that starting on 2022-05-09, over $5B of LUNA was traded on a single day, peaking at \\$12B on 2022-05-12.  Starting on 2022-05-11, the trading volume of **LUNA** was much higher than its market cap.  This is a strong indication that a large amount of intraday arbitrage trading involving **LUNA** was occuring. 


## Summary

During this project, a historic event occured related to stablecoin volatility.  I am referring, of course, to the collapse of US Terra and its counterpart Terra Luna.
UST is an algorithmic stablecoin and it has a special relationship with the Luna.  Luna and UST are swapped in order to maintain the UST at its peg value of \\$1.00 USD.

One of the striking observations we made when analyzing the collapse time frame is that the daily trading volumes were excessively high.  Late in the week, the trading volume of Luna was several times that of its market cap, strongly sugggesting that a massive amount of arbitrage was occuring at high frequency intraday.

This made us realize that there was a whole different type of arbitrage to consider than the one we modelled.
    Instead of arbitrage between the stablecoin's spot price and it's peg value spanning a few days, 
    for some algo coins like UST, the arbitrage is nearly instantaneous and between the stablecoin's spot price and other coins that are used to maintain the peg price.
